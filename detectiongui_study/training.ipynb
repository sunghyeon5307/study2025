```python
import yaml

data = {
    'train': '/content/dataset/train',
    'val': '/content/dataset/valid',
    'nc': 1,
    'names': ['penguin']
}

with open('/content/dataset/data.yaml', 'w') as f:
    yaml.dump(data, f)

with open('/content/dataset/data.yaml', 'r') as f:
    aquarium_yaml = yaml.safe_load(f)
    print(aquarium_yaml)  # display()는 Jupyter 전용이므로 print 사용
```

```python
import ultralytics

ultralytics.checks()

from ultralytics import YOLO

model = YOLO('yolov8s.pt')
```

```python
print(type(model.names), len(model.names))

print(model.names)
```

```python
model.train(
    data='/content/dataset/data.yaml',
    epochs=100,
    imgsz=640,                # 🔼 416보다 640이 일반적으로 더 나은 성능
    batch=32,                 # VRAM 여유 있으면 32까지 가능 (Colab GPU 기준 16~32)
    patience=50,              # 조기종료 안 하려면 100과 비슷하게
    workers=2,                # Colab에서는 2~4 추천
)
```

```python
print(type(model.names), len(model.names))
print(model.names)
```

```python
results = model.predict(source='/content/Aquarium_Data/test/images', save=True)
```
